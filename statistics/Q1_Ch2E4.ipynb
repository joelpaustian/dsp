```python
import nsfg
import pandas as pd
from IPython.display import Image
import matplotlib.pyplot as plt
import math as mt
%matplotlib inline
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 10)
pd.set_option('display.precision', 3)
preg=nsfg.ReadFemPreg()
live=preg[preg.outcome==1]
#live['totalwgt_lb'].value_counts().plot('hist')
live['totalwgt_lb']
dftw=pd.DataFrame(live['totalwgt_lb'])
plt.figure()
dftw.plot.hist(bins=100)
```

```python
live.head(5)
```

```python
firstbabies=live[live.birthord==1]
laterbabies=live[live.birthord!=1]
mean1=firstbabies['totalwgt_lb'].mean()
mean2=laterbabies['totalwgt_lb'].mean()
std1=firstbabies['totalwgt_lb'].std()
std2=laterbabies['totalwgt_lb'].std()
print(mean1,mean2,std1,std2)
var1=firstbabies['totalwgt_lb'].var()
var2=laterbabies['totalwgt_lb'].var()
n1, n2 = len(firstbabies['totalwgt_lb']), len(laterbabies['totalwgt_lb'])
pooled_var = (n1 * var1 + n2 * var2) / (n1 + n2)
d=(mean2-mean1)/mt.sqrt(pooled_var)
print(d) #Number of std dev's difference between groups
```

```python
#Repeat for prglngth
mean1=firstbabies['prglngth'].mean()
mean2=laterbabies['prglngth'].mean()
std1=firstbabies['prglngth'].std()
std2=laterbabies['prglngth'].std()
print(mean1,mean2,std1,std2)
var1=firstbabies['prglngth'].var()
var2=laterbabies['prglngth'].var()
n1, n2 = len(firstbabies['prglngth']), len(laterbabies['prglngth'])
pooled_var = (n1 * var1 + n2 * var2) / (n1 + n2)
d=(mean1-mean2)/mt.sqrt(pooled_var)
print(d) #Number of std dev's difference between groups
```
