```python
from __future__ import print_function, division

%matplotlib inline

import numpy as np

import nsfg
import first
import thinkstats2
import thinkplot
```

```python
randoms=np.random.random(1000)
```

```python
randoms_cdf = thinkstats2.Cdf(randoms)
thinkplot.Cdf(randoms_cdf)
thinkplot.Show(xlabel='random fraction of 1', ylabel='CDF')
```

```python
randoms_pmf=thinkstats2.Pmf(randoms)
thinkplot.Pmf(randoms_pmf)
thinkplot.Show(xlabel='random fraction of 1', ylabel='PMF')
```
